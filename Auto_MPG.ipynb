```python
import keras
import pandas as pd
from sklearn.model_selection import train_test_split
from keras import layers
from keras import Sequential
```

```python
data = pd.read_table('auto-mpg.data')
name = ['MPG', 'Cylinders', 'Displacement', 'Horsepower', 'Weight', 'Acceleration', 'Model Year', 'Origin']
data = pd.read_csv('auto-mpg.data', names = name, na_values = "?", comment ='\t', sep = " ", skipinitialspace = True)
data = data.dropna()
data = pd.get_dummies(data, columns = ['Origin'])
```

```python
traindata = data.drop(['MPG'], axis = 1)
label = data['MPG']
traindata_x, testdata_x, traindata_y, testdata_y = train_test_split(traindata, label, test_size = 0.3)
```

```python
traindata_1 = traindata_x.describe().transpose()
testdata_normalize = (testdata_x - traindata_1['mean']) / traindata_1['std']
traindata_normalize = (traindata_x - traindata_1['mean']) / traindata_1['std']
```

```python
model = Sequential([
    layers.Dense(128, activation = 'relu', input_shape=[len(traindata_normalize.keys())]),
    layers.Dense(128, activation = 'relu'),
    layers.Dense(1)
])
```

```python
model.compile(loss='MSE', optimizer = 'RMSprop', metrics = ['MAE', 'MSE'])
```

```python
summary = model.fit(traindata_normalize, traindata_y, epochs = 1000, validation_split = 0.3)
```

```python
table = pd.DataFrame(summary.history)
table['epoch'] = summary.epoch
table.tail()
```

```python
loss, MAE, MSE = model.evaluate(testdata_normalize, testdata_y)
print("The Mean Absolute Error is : {:2.5f} mpg".format(MAE))
```
